```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("EastWestAirlines.xlsx",sheet_name='data')
```

```python
df.head()
```

### EDA

```python
# droping the id column
```

```python
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=250
```

```python
df.drop(columns=['ID#'],inplace=True)
```

```python
df.head(2)
```

```python
# Taking Important column
```

```python
data=df.loc[:,['Balance','Flight_miles_12mo','Days_since_enroll','Bonus_miles']]
```

```python
data.head()
```

```python
data.describe()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x=sc.fit_transform(data)
x
```

### Clustering

```python
km=KMeans()
y=km.fit_predict
```

### Elbow Graph

```python
wcss=[]
for i in range(1,21):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
    
wcss
```

```python
plt.plot(range(1,21),wcss)
plt.xlabel('kvalue')
plt.ylabel('wcss value')
```

### Final model

```python
km=KMeans(n_clusters=5)
y=km.fit_predict(x)
y
```

```python
data['cluster']=y
```

```python
data.head()
```

```python
data.groupby('cluster').agg('mean').reset_index()
```

```python
km.cluster_centers_
```

```python
plt.scatter(x=data['Balance'],y=data['Flight_miles_12mo'],c=data['cluster'],cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],km.cluster_centers_[:,2],c='yellow')
```
