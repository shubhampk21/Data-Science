### Label Encoder

```python
columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
le=LabelEncoder()
```

```python
for i in columns:
    df_train[i]=le.fit_transform(df_train[i])
    df_test[i]=le.fit_transform(df_test[i])
```

```python
df_train.head()
```

```python
df_test.head()
```

### splitting the data

```python
colnames = df_train.columns
colnames

x_train = df_train[colnames[0:13]].values
y_train = df_train[colnames[13]].values
x_test = df_test[colnames[0:13]].values
y_test = df_test[colnames[13]].values
```

### Final Model

```python
nb=MultinomialNB()
nb.fit(x_train,y_train)
ypred=nb.predict(x_test)
```

```python
print("Training Accuracy",nb.score(x_train,y_train))
```

```python
print(classification_report(y_test,ypred))
```
